```python
# dependencies

# for df
import pandas as pd
import sklearn

# for tree visualization module 
from sklearn.tree import plot_tree


# for ml algo
# from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# for validation and accuracy score [0,1]
from sklearn.metrics import accuracy_score

#persisting models
import joblib

# music_data is a dataframe 
# .read_csv(pathname) to load csv files
# copy pathname from finder with right-click and holding option
music_data = pd.read_csv('/Users/shasvath/Desktop/Jupyter/ML_Exploration/Data/music.csv')
music_data
```

```python
# (rows, columns) of df 
music_data.shape

# basic info with respect to each common 
music_data.describe()

# show arrays in numpy rep with respect to df 
music_data.values
```

```python
# Training sets 

# input set often referenced as X
# .drop removes specified column
X = music_data.drop(columns = ['genre'])

# output set often refernced as y 
# only contains answers or predictions 
y = music_data['genre']

# ––––––––––

# Validation

# we must also test the accuracy of our model
# generally we allocate 70-80% of data to training and 20-30% for validation
# here we allocate 20%
# returns a tuple
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)


# ––––––––––
 
# predictions algo

# new instance of class
model = DecisionTreeClassifier()

# learn patterns from data (input, output)
model.fit(X_train.values, y_train)

# ––––––––––
# visualize decision tree
sklearn.tree.plot_tree(model)


# ––––––––––

# persisting model implementation

# store trained model 
# joblib.dump(model, "music-rec.joblib")

# call stored model
# model = joblib.load(model, "music-rec.joblib")

# ––––––––––

# prediction query

# .predict takes in parameters to return an output
# here one thing we ask is to predict "[21,1]"; As per our initial dataset this alludes to a 21(21) year old male(1)
# predictions = model.predict([ [21,1],[22,1] ])
# predictions

# ––––––––––

# accuracy score 
# predictions = model.predict(X_test.values)
# score = accuracy_score(y_test, predictions)
# score
```

<br><br>
As problems become more complex, more data is needed to come to a definitive solution. This is where persisting models come in.
